```python
import random
import string
import nltk
```

```python
# Descargar la lista de palabras en inglés si aún no está descargada
nltk.download('words')

# Cargar la lista de palabras en inglés
from nltk.corpus import words

# Obtener todas las palabras en inglés
word_list = words.words()

# Generar una lista de 500 palabras aleatorias
palabras_aleatorias = random.sample(word_list, 500)

# Imprimir las palabras aleatorias
for palabra in palabras_aleatorias:
    print(palabra)
```

```python
def merge_sort(arr):
    if len(arr) > 1:
        # Split the array into two halves
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        # Recursive calls to sort both halves
        merge_sort(left_half)
        merge_sort(right_half)

        # Initialize indices for traversing the two halves and the main array
        i = j = k = 0

        # Merge the two halves back into the original array
        while i < len(left_half) and j < len(right_half):
            if left_half[i] <= right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        # Check if any elements were left in either of the halves
        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

    return arr
```

```python
def lista(lista, desde, hasta):
    lista_rango = []
    for i in range(desde, hasta):
        lista_rango.append((lista[i]).lower())
    return merge_sort(lista_rango)
```

```python
lista(palabras_aleatorias, 0, 10)
```

```python
lista = lista(palabras_aleatorias, 0, 10)
```
